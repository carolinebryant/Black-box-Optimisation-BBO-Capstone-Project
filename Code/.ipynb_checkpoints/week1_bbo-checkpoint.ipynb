# Helper Functions

```python
def compute_ucb(mu, sigma, kappa=2.5):
    """Compute Upper Confidence Bound acquisition function"""
    return mu + kappa * sigma

def compute_ei(mu, sigma, f_best, xi=0.01):
    """Compute Expected Improvement acquisition function"""
    z = (mu - f_best - xi) / (sigma + 1e-9)
    ei = (mu - f_best - xi) * norm.cdf(z) + sigma * norm.pdf(z)
    ei[sigma == 0] = 0
    return ei

def format_query(point, decimals=6):
    """Format point for portal submission"""
    point_clipped = np.clip(point, 0, 0.999999)
    return '-'.join([f'{x:.{decimals}f}' for x in point_clipped])

def analyze_improvement(outputs):
    """Analyze if last query improved over previous best"""
    if len(outputs) < 2:
        return None, None
    last_output = outputs[-1]
    previous_best = outputs[:-1].max()
    improvement = last_output - previous_best
    return improvement, last_output >= previous_best
```

# Function 7: ML Hyperparameters (6D)

```python
# Load function 7 data
f7_inputs = np.load('../Data/function_7/initial_inputs.npy')
f7_outputs = np.load('../Data/function_7/initial_outputs.npy')

print("X shape:", f7_inputs.shape)
print("Y shape:", f7_outputs.shape)
print("\nX:")
print(f7_inputs)
print("\nY values:")
print(f7_outputs)
```

```python
print(f"Function 7 - Week 1")
print(f"Data shape: {f7_inputs.shape}")
print(f"Output range: [{f7_outputs.min():.6f}, {f7_outputs.max():.6f}]")
print(f"Current best: {f7_outputs.max():.6f}")

# Analyze dimension importance
best_quartile = f7_outputs >= np.percentile(f7_outputs, 75)
worst_quartile = f7_outputs <= np.percentile(f7_outputs, 25)

print("\nDimension correlations:")
for dim in range(6):
    top_mean = f7_inputs[best_quartile, dim].mean()
    bottom_mean = f7_inputs[worst_quartile, dim].mean()
    diff = abs(top_mean - bottom_mean)
    print(f"  x{dim+1}: diff={diff:.3f}")

# Fit GP
kernel_f7 = C(1.0, (1e-3, 1e3)) * Matern(length_scale=0.5, length_scale_bounds=(0.1, 2.0), nu=2.5)
gp_f7 = GaussianProcessRegressor(kernel=kernel_f7, n_restarts_optimizer=15, alpha=1e-6, normalize_y=True)
gp_f7.fit(f7_inputs, f7_outputs)

print(f"\nGP kernel: {gp_f7.kernel_}")

# Generate candidates (random + biased)
np.random.seed(42)
random_f7 = np.random.uniform(0, 1, (10000, 6))

biased_f7 = np.zeros((10000, 6))
for dim in range(6):
    top_values = f7_inputs[best_quartile, dim]
    mean_top = top_values.mean()
    std_top = top_values.std()
    biased_f7[:, dim] = np.random.normal(mean_top, std_top * 1.5, 10000)
    biased_f7[:, dim] = np.clip(biased_f7[:, dim], 0, 1)

candidates_f7 = np.vstack([random_f7, biased_f7])

# Predict and compute UCB
mu_f7, sigma_f7 = gp_f7.predict(candidates_f7, return_std=True)
ucb_f7 = compute_ucb(mu_f7, sigma_f7, kappa=2.5)

# Select best
best_idx_f7 = np.argmax(ucb_f7)
query_f7_week1 = candidates_f7[best_idx_f7]

print(f"\nWeek 1 Query: {format_query(query_f7_week1)}")
print(f"Predicted μ: {mu_f7[best_idx_f7]:.6f}, σ: {sigma_f7[best_idx_f7]:.6f}")
print(f"UCB: {ucb_f7[best_idx_f7]:.6f}")
```

# Function 8: High-Dimensional ML (8D)

```python
# Load function 8 data
f8_inputs = np.load('../Data/function_8/initial_inputs.npy')
f8_outputs = np.load('../Data/function_8/initial_outputs.npy')

print("X shape:", f8_inputs.shape)
print("Y shape:", f8_outputs.shape)
print("\nX:")
print(f8_inputs)
print("\nY values:")
print(f8_outputs)
```

```python
print(f"Function 8 - Week 1")
print(f"Data shape: {f8_inputs.shape}")
print(f"Output range: [{f8_outputs.min():.6f}, {f8_outputs.max():.6f}]")
print(f"Current best: {f8_outputs.max():.6f}")

# Analyze dimension importance
best_quartile = f8_outputs >= np.percentile(f8_outputs, 75)
worst_quartile = f8_outputs <= np.percentile(f8_outputs, 25)

print("\nDimension correlations:")
for dim in range(8):
    top_mean = f8_inputs[best_quartile, dim].mean()
    bottom_mean = f8_inputs[worst_quartile, dim].mean()
    diff = abs(top_mean - bottom_mean)
    print(f"  x{dim+1}: diff={diff:.3f}")

# Fit GP
kernel_f8 = C(1.0, (1e-3, 1e3)) * Matern(length_scale=0.5, length_scale_bounds=(0.1, 2.0), nu=2.5)
gp_f8 = GaussianProcessRegressor(kernel=kernel_f8, n_restarts_optimizer=15, alpha=1e-6, normalize_y=True)
gp_f8.fit(f8_inputs, f8_outputs)

print(f"\nGP kernel: {gp_f8.kernel_}")

# Generate candidates (random + biased + local)
np.random.seed(42)
random_f8 = np.random.uniform(0, 1, (15000, 8))

biased_f8 = np.zeros((15000, 8))
for dim in range(8):
    top_values = f8_inputs[best_quartile, dim]
    mean_top = top_values.mean()
    std_top = top_values.std()
    biased_f8[:, dim] = np.random.normal(mean_top, std_top * 1.5, 15000)
    biased_f8[:, dim] = np.clip(biased_f8[:, dim], 0, 1)

top5_f8 = np.argsort(f8_outputs)[-5:][::-1]
local_f8 = []
for idx in top5_f8:
    for _ in range(1000):
        perturb = np.random.normal(0, 0.08, 8)
        candidate = np.clip(f8_inputs[idx] + perturb, 0, 1)
        local_f8.append(candidate)

candidates_f8 = np.vstack([random_f8, biased_f8, np.array(local_f8)])

# Predict and compute UCB
mu_f8, sigma_f8 = gp_f8.predict(candidates_f8, return_std=True)
ucb_f8 = compute_ucb(mu_f8, sigma_f8, kappa=2.5)

# Select best
best_idx_f8 = np.argmax(ucb_f8)
query_f8_week1 = candidates_f8[best_idx_f8]

print(f"\nWeek 1 Query: {format_query(query_f8_week1)}")
print(f"Predicted μ: {mu_f8[best_idx_f8]:.6f}, σ: {sigma_f8[best_idx_f8]:.6f}")
print(f"UCB: {ucb_f8[best_idx_f8]:.6f}")
```
